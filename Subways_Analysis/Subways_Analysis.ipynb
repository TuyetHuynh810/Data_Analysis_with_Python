```python
# calculate correlation of two variables 
# step 1: standardize ( in standard deviations )
# step 2: average of X in standard units and Y in standard units 
data.rain.value_counts()
```

```python
# use these for correlation calculations 
entries = data['ENTRIESn_hourly']
cum_entries = data['ENTRIESn']
rain = data['meanprecipi']
temp = data['meantempi']
```

```python
def correlation(x, y):
    '''
    Fill in this function to compute the correlation between the two
    input variables. Each input is either a NumPy array or a Pandas
    Series.
    
    correlation = average of (x in standard units) times (y in standard units)
    
    Remember to pass the argument "ddof=0" to the Pandas std() function!
    '''
    
    # first standardize the variables 
    x_nor = ( x - x.mean()) / x.std(ddof = 0)
    y_nor = ( y - y.mean()) / y.std(ddof = 0)

    correlation = (x_nor * y_nor).sum() / x_nor.shape[0]
    return correlation
```

```python
rain.std()
rain.mean()
# rain z -scores 
rain_z = (rain - rain.mean())/rain.std(ddof = 0)
```

```python
entries_z = (entries - entries.mean())/entries.std(ddof = 0)
```

```python
r = (rain_z * entries_z).sum() / rain_z.shape[0]
r
```

```python
correlation(entries, rain)
```

```python
correlation(entries, temp)
```

```python
correlation(rain, temp)
```
